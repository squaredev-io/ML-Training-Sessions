## **3. Basic types of Machine Learning**

The basic types of Machine Learning are supervised learning, semi-supervised learning, unsupervised learning, and reinforcement learning.

**3.1   Supervised Learning**

This type of Machine Learning includes training a program on labeled data, where both the input (Xi) and result (Yi) are known. The program figures out how to map inputs to outputs by reducing the difference between what it thinks will happen and what actually happens.

Each one of the Xi set is called a **"feature vector"**. A vector is actually an 1-dimentional array. The length of this array is called **"dimentionality"**.

Example: In case that we need to find if emails are spam, or not_spam, we need to predict the result. This is called classification as we have to predict a class (spam, not_spam). 
In another example if we had to predict the height of a person (a real number) we would apply a regression solving. In this case the y (output) is called **"target"**.

Types of supervised learning algorithms are:
- Nearest Neighbor
- Naive Bayes
- Decision Trees
- Linear Regression
- Support Vector Machines (SVM)
- Neural Networks

The goal of a supervised learning method is to use a dataset to make a model that takes a feature vector x as input and returns information that lets you figure out a label for this feature vector. For example, a model built from a set of data about patients could take as input a "feature vector" that describes a patient and output the likelihood that the patient has cancer.

**3.2   Unsupervised Learning**

In unsupervised learning, the algorithm is trained on unlabeled data and must find patterns or relationships on its own. The goal of unsupervised learning is to uncover the underlying structure or distribution of the data. Clustering and anomaly detection are some common examples of unsupervised learning.

Types of unsupervised algorithms
- K-Means Clustering
- Principal Component Analysis
- Hierarchical Clustering



**More information**

Clustering is a good way to find groups of similar images or written documents in a big set of items. By using clustering, for example, the analyst can pick a small but representative subset of unlabeled examples from a large set of examples for manual labeling. Instead of picking examples directly from the large set and risking picking only very similar examples, a few examples are picked from each cluster.

In dimensionality reduction, the result of the model is a feature vector with fewer dimensions than the input. For example, the expert has a feature vector that has more than three dimensions and is too hard to picture. The dimensionality reduction model can take that feature vector and turn it into a new feature vector with only two or three dimensions, while keeping some of the information. A graph can be made with this new feature vector.

In outlier detection, the result is a real number that shows how the input feature vector is different from a "typical" case in the dataset. Outlier detection can help solve a network attack problem by finding abnormal network packets that are different from a typical packet in "normal" traffic. It can also be used to find something new, like a document that is different from the others in a collection.

**3.3   Semi-supervised Learning**

In semi-supervised learning, both labeled and unlabeled data exist in the collection. Usually, there is a lot more data that doesn't have labels than data that do. The goal here is for a program that learns from data to find (or "produce" or "compute") a better model by looking at many data without labels.


**3.4   Reinforcement Learning**

Reinforcement learning is a subfield of machine learning in which the machine (called an agent) "lives" in an environment and can see the state of that environment as a vector of features. In non-terminal states, the machine can do things. Different acts give different rewards and could also move the machine to a different part of the environment. A common goal of a program for reinforcement learning is to learn the best thing to do.

An optimal policy is a function (like the model in supervised learning) that takes as input the feature vector of a state and returns the best action to take in that state. The move is best if it increases the average estimated long-term gain by the most.

Reinforcement learning is a way to solve a problem where decisions are made in a certain order and the goal is long-term, like in games, robots, managing resources, or transportation.



## **4.   Main data concepts (structured, unstructured, and time-series)**

**4.1 Structured Data**

Structured Data: Structured data is organized in a very organized and structured way, usually in a tabular style like a spreadsheet or a database table. Structured data is very well organized and follows a strict data model or schema. Sales data, customer data, and product data are all types of organized data.


**4.2 Unstructured Data**

Text, pictures, audio, and video data are all examples of unstructured data because they are not set up in a certain way. Unstructured data, on the other hand, does not have a model, and the links between data pieces are not well-defined. Unstructured data includes things like posts on social media, customer reviews, and data from sensors.

**4.3 Timeseries Data**

Timeseries is mainly consisted of structured data.
Data that is collected at regular times over time is called "time-series data." Time-series statistics are often used to model and study long-term trends and patterns. Stock prices, weather data, and website traffic data are all types of time-series data.

Structured data

- Quantitative (numerical)

- Easily searchable

Unstructured data

- Qualitative (non-numerical)
- Highly disorganized and subjective

- Not organized using predefined formats or data models
- In the form of disordered texts, images, videos, sounds
- Extremely difficult to search

## **5.   Briefly description of data preprocessing**

Data preparation and preprocessing is an important step in any Machine Learning project. It includes cleaning, changing, and preparing the data to be used to build a model. Here are some steps that are often taken to prepare data:

- **Data cleaning** is the process of finding any lost, wrong, or useless data and fixing it or getting rid of it.

- **Data transformation** is the process of putting the data into a shape that can be used for research and modeling. This can be done by scaling or standardizing the data, for example.

- **Feature Extraction** is the process of making new features from the current data, such as by using methods like principal component analysis (PCA) to reduce the number of dimensions in the data.

- **Data integration** is the process of putting together information from different sources into a single dataset.

- **Data Reduction** is done by choosing or selecting the data to cut down on the size of the dataset.

- **Data Discretization** includes turning continuous data into defined groups.

- **Data augmentation** is a way to increase the size of the training set by making changes to copies of an existing dataset. It can involve making small changes to the information or using deep learning to create new data points.

Most of the time, these steps of preparation need to be done more than once in order to get the best results. The goal of data preparation is to improve the quality and usefulness of the data, cut down on noise and duplication, and make the model more accurate and better at what it does.

### **6.1 numpy** 

```python
import numpy as np
```

**Creating arrays**

You can create a numpy array by passing a Python list to the np.array() function. For example:

```python
# create a 1-dimensional array
a = np.array([1, 2, 3])
print(a)
print("-----------")
# create a 2-dimensional array
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)
```

**Basic Operations**

numpy provides many built-in functions for performing operations on arrays. For example:

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# addition
c = a + b
print(c)
print("-----")
# multiplication
d = a * b
print(d)
```

**Indexing and Slicing**

You can access and modify elements of a numpy array using indexing and slicing. For example:

```python
# get the first element
print(a[0])
print("-----")
# get the last element
print(a[-1])
print("-----")
# get a slice of the array
print(a[1:3])
print("-----")
# modify an element
a[0] = 10
```

These are just a few basic examples of using NumPy in Python. NumPy provides many more advanced features and functions for numerical computing, such as linear algebra operations, statistical analysis, and more.

### **6.2 pandas** 

```python
import pandas
```

**Creating DataFrames**

You can create a Pandas DataFrame by passing a dictionary or a list of lists to the pd.DataFrame() function. For example:

```python
# create a DataFrame from a dictionary
data = {'name': ['Alice', 'Bob', 'Charlie'],
        'age': [25, 30, 35],
        'gender': ['F', 'M', 'M']}
df = pd.DataFrame(data)

# create a DataFrame from a list of lists
data = [['Alice', 25, 'F'], ['Bob', 30, 'M'], ['Charlie', 35, 'M']]
df = pd.DataFrame(data, columns=['name', 'age', 'gender'])
display(df)
```

**Basic Operations**

Pandas provides many built-in functions for performing operations on DataFrames. For example:

```python
# get the first 5 rows
display(df.head())
```

```python
# get a summary of the data
display(df.describe())
```

```python
# group the data by gender
display(df.groupby('gender').mean())
```

**Indexing and Slicing**

You can access and modify elements of a Pandas DataFrame using indexing and slicing. For example:

```python
# get the first element of the 'name' column
display(df.loc[0, 'name'])
```

```python
# get a slice of the DataFrame
print(df.iloc[1:3, :])
```

```python
display(df)
print("-----")
# modify an element
df.loc[0, 'age'] = 30
display(df)
```

These are just a few basic examples of using Pandas in Python. Pandas provides many more advanced features and functions for data manipulation and analysis, such as merging, filtering, reshaping, and more.

### **6.4 pillow** 

```python
from PIL import Image, ImageFilter, ImageDraw, ImageFont
```

Loading and Displaying Images

You can load an image using the Image.open() function and display it using the Image.show() function. For example:

```python
# load an image
img = Image.open('/content/AI_ML_AI.png')

# display the image
img.show()
```

Image Processing

Pillow provides many built-in functions for image processing, such as resizing, cropping, and filtering. For example:

```python
# resize the image
resized = img.resize((500, 500))

# crop the image
cropped = img.crop((100, 100, 400, 400))

# apply a Gaussian blur
blurred = img.filter(ImageFilter.GaussianBlur(radius=5))

# display the images
img.show()
resized.show()
cropped.show()
blurred.show()
```

**Image Manipulation**

You can manipulate images using the ImageDraw and ImageFont modules. For example:

```python
# create an ImageDraw object
draw = ImageDraw.Draw(img)

# draw a rectangle on the image
draw.rectangle((100, 100, 400, 400), outline='red', width=2)
```

```python
# create an ImageDraw object
draw = ImageDraw.Draw(img)

# draw a rectangle on the image
draw.rectangle((100, 100, 400, 400), outline='red', width=2)

# add text to the image
font = ImageFont.load_default()
draw.text((50, 50), 'Hello, world!', fill='blue', font=font)

# display the image
img.show()
```

These are just a few basic examples of using Pillow in Python. Pillow provides many more advanced features and functions for image processing and manipulation, such as image enhancement, color manipulation, and image file format conversion.
